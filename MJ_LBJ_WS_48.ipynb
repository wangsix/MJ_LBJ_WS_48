```python
from basketball_reference_scraper.players import get_stats, get_game_logs, get_player_headshot
from basketball_reference_scraper.teams import get_team_misc, get_roster_stats, get_team_stats
```

```python
MJ_matchup_list = [
    {"year": 1991, "a":"CHI", "b":"LAL"},
    {"year": 1992, "a":"CHI", "b":"POR"},
    {"year": 1993, "a":"CHI", "b":"PHO"},
    {"year": 1996, "a":"CHI", "b":"SEA"},
    {"year": 1997, "a":"CHI", "b":"UTA"},
    {"year": 1998, "a":"CHI", "b":"UTA"},
]
```

```python
team_list = []
oppo_list = []
for match_up in MJ_matchup_list:
    team_df = get_roster_stats(match_up["a"], match_up["year"], "ADVANCED", True)
    team_df['MPG'] = team_df.apply(lambda row: float(row['MP'])/float(row['G']), axis=1)
    team_ws_df = team_df.loc[team_df['MPG']>=15.0].loc[:, ['PLAYER', 'WS/48', 'SEASON', 'MP']].sort_values("WS/48", ascending=False)
    team_list.append(team_ws_df)
    print(team_ws_df)
    oppo_df = get_roster_stats(match_up["b"], match_up["year"], "ADVANCED", True)
    oppo_df['MPG'] = oppo_df.apply(lambda row: float(row['MP'])/float(row['G']), axis=1)
    oppo_ws_df = oppo_df.loc[oppo_df['MPG']>=15.0].loc[:, ['PLAYER', 'WS/48', 'SEASON', 'MP']].sort_values("WS/48", ascending=False)
    oppo_list.append(oppo_ws_df)
    print(oppo_ws_df)
```

```python
import pandas as pd
```

```python
def aggregate_func(x):
    d = {}
    d['WS/48 mean'] = x['WS/48'].mean()
    d['MP mean'] = x['MP'].mean()
    d['SEASON count'] = x['SEASON'].count()
    d['WS total'] = (x['WS/48'] * x['MP']).sum()/x['MP'].sum()
    return pd.Series(d, index=['WS/48 mean', 'MP mean', 'SEASON count', 'WS total'])

```

```python
mj_df = pd.concat(team_list).sort_values("WS/48", ascending=False).reset_index(drop=True)
mj_df['WS/48'] = mj_df['WS/48'].astype(float)
mj_df['MP'] = mj_df['MP'].astype(float)
```

```python
mj_df
```

```python
mj_avg_ws_df = mj_df.groupby('PLAYER').apply(aggregate_func).sort_values(('WS total'), ascending=False).reset_index()
```

```python
mj_avg_ws_df
```

```python
mj_teammate_ws_df = mj_df.loc[mj_df['PLAYER']!='Michael Jordan']
```

```python
(mj_teammate_ws_df['WS/48']*mj_teammate_ws_df['MP']).sum()/mj_teammate_ws_df['MP'].sum()
```

```python
mj_oppo_df = pd.concat(oppo_list).sort_values("WS/48", ascending=False).reset_index(drop=True)
mj_oppo_df['WS/48'] = mj_oppo_df['WS/48'].astype(float)
mj_oppo_df['MP'] = mj_oppo_df['MP'].astype(float)
```

```python
mj_oppo_avg_ws_df = mj_oppo_df.groupby('PLAYER').apply(aggregate_func).sort_values(('WS total'), ascending=False).reset_index()
```

```python
mj_oppo_avg_ws_df
```

```python
(mj_oppo_df['WS/48']*mj_oppo_df['MP']).sum()/mj_oppo_df['MP'].sum()
```
